```
# images using skikit-image 
import os  
# importing io from skimage 
import skimage 
from skimage import io 
from skimage import data, data_dir
# way to load car image from file 
file = os.path.join(skimage.data_dir, 'astronaut.png') 
photo = io.imread(file) 
# way to show the input image 
io.imshow(photo) 
io.show()
```
